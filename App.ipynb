```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from state_choices import STATE_CHOICES

#load the dataset 
data_path = './Metro_invt_fs_uc_sfrcondo_sm_month.csv'
df = pd.read_csv(data_path)
df.head()
```

```python
# Transpose the dataframe to have dates as rows for easier plotting
# We will also filter for a few regions to make the plots more readable
regions_to_plot = ['United States', 'New York, NY', 'Los Angeles, CA', 'Chicago, IL', 'Dallas, TX']
df_filtered = df[df['StateName'].isin(STATE_CHOICES)]
df_transposed = df_filtered.set_index('StateName').T

# Reset the index to have dates as a column
df_transposed.reset_index(inplace=True)
df_transposed.rename(columns={'index': 'Date'}, inplace=True)

# Convert the Date column to datetime format
df_transposed['Date'] = pd.to_datetime(df_transposed['Date'], errors='coerce')

# Drop rows with NaT in the Date column
df_transposed.dropna(subset=['Date'], inplace=True)

df_transposed
```
